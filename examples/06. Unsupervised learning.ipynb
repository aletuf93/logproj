### Import data

```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import sys; sys.path.insert(0, '..') #add the above level with the package
from logproj import ml_dimensionalityReduction as dr
from logproj import ml_unsupervised_models as uns_mod



#import data
from sklearn.datasets import load_digits
#Load the dataset
data = load_digits()

# define X dataframe
X = data.data
X = pd.DataFrame(X)

# define y dataframe
y = data.target
y = pd.DataFrame(y,columns=['target'])

```

### K-means

```python
#project the dataset X into two component to allow plotting
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=2,XX=X, diagnose=True)


inputColumns = X.columns
X_clustered =  uns_mod.groupVariableKMean(X,inputColumns,k=10)
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=2,XX=X, diagnose=True)

plt.figure()
plt.scatter(X_res[0],X_res[1],c = X_clustered['CLUSTER_KMEANS_10'], cmap=cm.plasma)
plt.title('K-means clustering')

plt.figure()
plt.scatter(X_res[0],X_res[1],c = y['target'], cmap=cm.plasma)
plt.title('True labels')

```

### Gaussian Mixture Models

```python
X_clustered =  uns_mod.GroupingVariableGMM(X,inputColumns,k=10)
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=2,XX=X, diagnose=True)

plt.figure()
plt.scatter(X_res[0],X_res[1],c = X_clustered['CLUSTER_GMM_10'], cmap=cm.plasma)
plt.title('Gaussian Mixture models clustering')

plt.figure()
plt.scatter(X_res[0],X_res[1],c = y['target'], cmap=cm.plasma)
plt.title('True labels')
```

### Hierarchical clustering

```python
#Dendrogram using single linkage
out = uns_mod.HierarchicalClusteringDendrogram(X,
                                       metodoGrouping='single',
                                       medotoDistanze='euclidean')
out['dendrogram']
```

```python
#Dendrogram using complete linkage
out = uns_mod.HierarchicalClusteringDendrogram(X,
                                       metodoGrouping='complete',
                                       medotoDistanze='euclidean')
out['dendrogram']
```

```python
#Dendrogram using complete linkage
out = uns_mod.HierarchicalClusteringDendrogram(X,
                                       metodoGrouping='average',
                                       medotoDistanze='euclidean')
out['dendrogram']
```

```python
#Cut the dendrogram to define a fixed number of clusters k
#single linkage
X_clustered =  uns_mod.GroupingVariableHierarchical(X,inputColumns,k=10,metodoGrouping='single')
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=2,XX=X, diagnose=True)

plt.figure()
plt.scatter(X_res[0],X_res[1],c = X_clustered['CLUSTER_HIER_10'], cmap=cm.plasma)
plt.title('Single linkage clustering')

plt.figure()
plt.scatter(X_res[0],X_res[1],c = y['target'], cmap=cm.plasma)
plt.title('True labels')
```

```python
#Cut the dendrogram to define a fixed number of clusters k
#complete linkage
X_clustered =  uns_mod.GroupingVariableHierarchical(X,inputColumns,k=10,metodoGrouping='complete')
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=2,XX=X, diagnose=True)

plt.figure()
plt.scatter(X_res[0],X_res[1],c = X_clustered['CLUSTER_HIER_10'], cmap=cm.plasma)
plt.title('Complete linkage clustering')

plt.figure()
plt.scatter(X_res[0],X_res[1],c = y['target'], cmap=cm.plasma)
plt.title('True labels')
```

```python
#Cut the dendrogram to define a fixed number of clusters k
#average linkage
X_clustered =  uns_mod.GroupingVariableHierarchical(X,inputColumns,k=10,metodoGrouping='average')
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=2,XX=X, diagnose=True)

plt.figure()
plt.scatter(X_res[0],X_res[1],c = X_clustered['CLUSTER_HIER_10'], cmap=cm.plasma)
plt.title('Average linkage clustering')

plt.figure()
plt.scatter(X_res[0],X_res[1],c = y['target'], cmap=cm.plasma)
plt.title('True labels')
```
